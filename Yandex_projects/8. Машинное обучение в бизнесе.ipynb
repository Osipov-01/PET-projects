–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å —Ç–∞–±–ª–∏—Ü—É –±–µ–∑ –≤—ã–±—Ä–æ—Å–æ–≤, –∑–∞—Ç–µ–º –ø–æ–¥–ø—Ä–∞–≤–∏–º –¥–∞–Ω–Ω—ã–µ —Å –ø–æ–º–æ—â—å—é –≤—ã–∑–æ–≤–∞ —ç—Ç–æ–π —Å–∞–º–æ–π —Ñ—É–Ω–∫—Ü–∏–∏

```python
def df_process(df, columns):
    df_processed = df
    for column in columns:
        q25 = df_processed[column].quantile(.25)
        q75 = df_processed[column].quantile(.75)        
        q_min = q25 - 1.5*(q75-q25)
        q_max = q75 + 1.5*(q75-q25)
        df_processed = df_processed[df_processed[column] >= q_min]
        df_processed = df_processed[df_processed[column] <= q_max]
    return df_processed
        
```

```python
def income (target, pred):
    pred_sort = pd.Series(pred).sort_values(ascending=False)
    best_pred_sort = pred_sort[:selected_points]
    best_target_sort = target.reset_index(drop=True)[best_pred_sort.index]
    sum_target = best_target_sort.sum()
    income = sum_target * price_for_every_product - budget_region
    return round(income, 0)
```

–ë—ã–ª–æ –±—ã –∑–¥–æ—Ä–æ–≤–æ –∑–Ω–∞—Ç—å –∑–∞—Ä–∞–Ω–µ–µ 200 –ª—É—á—à–∏—Ö —Å–∫–≤–∞–∂–∏–Ω –∏–∑ 25_000 :) –ù–æ –ø–æ —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏ –º–æ–∂–µ–º –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ 500 —Ç–æ—á–µ–∫.

–†–∞—Å—á—ë—Ç —Ä–∏—Å–∫–æ–≤

```python
def confidence_interval(target, pred):
    incomes = []
    for i in range(1000):
        pred_sample = pd.Series(pred).sample(n = max_points, replace=True, random_state = state)
        incomes.append(income(target, pred_sample))
    incomes = pd.Series(incomes)
    print(f'–°—Ä–µ–¥–Ω—è—è –ø—Ä–∏–±—ã–ª—å: {incomes.mean()}')
    risk = len(incomes[incomes<0])/len(incomes) 
    print(f'–†–∏—Å–∫–∏: {"{:.0%}".format(risk)}')
    
    lower = incomes.quantile(0.025)
    upper = incomes.quantile(0.975)    
    
    return lower, upper
    
```
